```python
import os
# step 1 file and folder option creation
def create_file(filename):
  try:
    with open(filename ,'x'):
     print(f'{filename} is created sucessfully')
  except FileExistsError:
    print(f'{filename} already exists')
  except Exception as e:
    print(f'{filename} has some error during creation')

def create_folder(foldername):
  try:
    os.mkdir(foldername)
    print(f'{foldername} is created sucessfully')
  except FileExistsError:
    print(f'{foldername} already exists')
  except Exception as e:
    print(f'{foldername} has some error during creation')

def delete_file(filename):
  try:
    os.remove(filename)
    print(f'{filename} is deleted successfully')
  except FileNotFoundError:
    print(f'{filename} does not exist')
  except Exception as e:
    print(f'{filename} has some error during deletion')

def delete_folder(foldername):
  try:
    os.rmdir(foldername)
    print(f'{foldername} is deleted successfully')
  except FileNotFoundError:
    print(f'{foldername} does not exist')
  except Exception as e:
    print(f'{foldername} has some error during deletion')

def view_files():
  files =os.listdir()
  print('Files in current directory are:')
  for file in files:
    print(file)

def edit_file(filename):
  try:
    with open(filename , 'a') as f:
      content = input("Enter the content you want to add or edit")
      f.write(content + '\n')
      print(f'{filename} is edited successfully')
  except FileNotFoundError:
    print(f'{filename} does not exist')
  except Exception as e:
    print(f'{filename} has some error during editing')

def read_file(filename):
  try:
   with open(filename, 'r') as f:
    content = f.read()
    print(f'{filename} read successfully and content is: \n {content}')
  except FileNotFoundError:
    print(f'{filename} does not exist')
  except Exception as e:
    print(f'{filename} has some error during reading')

def current_folder():
  try:
    print(f"Current working directory: {os.getcwd()}")
  except Exception as e:
    print(f'Current directory has some issues')

def folder_change(foldername):
  try:
    os.chdir(foldername)
    print(f"Current working directory after chdir: {os.getcwd()}")
  except FileNotFoundError:
    print(f'{foldername} does not exist')
  except Exception as e:
      print(f'{foldername} has some error during changing')

def backward_folder():
  os.chdir('..')
  print(f"Current working directory after chdir: {os.getcwd()}")
#step 2 option display
def main():
  while True:
    print('"File & Folder Manager App"' )
    print('1. Create File')
    print('2. Read File')
    print('3. Edit File')
    print('4. Delete File')
    print('5. View All File In Currrent Folder/Directory')
    print('6. Create Folder')
    print('7 Show Current Folder')
    print('8. Change Folder')
    print('9. Backward Folder')
    print('10. Delete Folder')
    print('11. Exit')

    #step 3 logic
    choice = input('Enter your choice: ')
    if choice =='1':
      filename = input('Enter the File Name: ')
      create_file(filename)
    elif choice =='2':
      filename = input('Enter the File Name: ')
      read_file(filename)
    elif choice =='3':
      filename = input('Enter the File Name: ')
      edit_file(filename)
    elif choice =='4':
      filename = input('Enter the File Name: ')
      delete_file(filename)
    elif choice =='5':
      view_files()
    elif choice =='6':
      foldername = input('Enter the Folder Name: ')
      create_folder(foldername)
    elif choice =='7':
      current_folder()
    elif choice =='8':
      foldername = input('Enter the Folder Name: ')
      folder_change(foldername)
    elif choice =='9':
      backward_folder()
    elif choice =='10':
      foldername = input('Enter the Folder Name: ')
      delete_folder(foldername)
    elif choice =='11':
      print('Thank you for using File & Folder Manager App')
      break
    else:
      print('Invalid choice ')

if __name__ =='__main__':
  main()
```
